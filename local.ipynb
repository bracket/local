```python
import tasks
import matplotlib.pyplot as plt
import numpy as np

from datetime import date, timedelta, datetime
```

```python
from math import exp

def sigmoid(alpha, x):
    e = exp(alpha * x)
    return e / (1 + e)
```

```python
xs = np.linspace(-1, 1, num=100, endpoint=True, dtype=np.float64)
ys = np.array([ sigmoid(10, x) for x in xs ])
```

```python
plt.plot(xs, ys)
plt.show()
```
